```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
music_data = pd.read_csv('music.csv')
X = music_data.drop(columns = ['genre'])
y = music_data['genre']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

modal = DecisionTreeClassifier()
modal.fit(X_train,y_train)
prediction = modal.predict(X_test)

score = accuracy_score(y_test, prediction)
score
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random 

G=nx.Graph()
city_set = ['Mumbai', 'Chennai', 'Pune', 'Banglore', 'Surat', 'Hyderabad', 'Jammu', 'Kashmir']
for each in city_set:
    G.add_node(each)
    
costs=[]
value=100

while(value<=2000):
    costs.append(value)
    value=value+100

while(G.number_of_edges()<16):
    c1=random.choice(list(G.nodes()))
    c2=random.choice(list(G.nodes()))
    if c1!=c2 and G.has_edge(c1,c2) == 0:
        w=random.choice(costs)
        G.add_edge(c1,c2,weight=w)
        
pos=nx.circular_layout(G)
nx.draw(G,with_labels=True)
plt.show()
```
